# Max Length Tokenizer Algorithm
Here we will implement a Max Length Tokenizer for the both farsi and english languages.

```python
import threading
```

```python
def max_length_tokenizer(merged, tokens):
    start = 0
    splited_tokens = []
    
    available_tokens = tokens
    possible_choices = []
    for i in range(len(merged)):
        available_tokens = list(filter( lambda x : x.startswith(merged[start:i+1]), available_tokens))
        for token in available_tokens:
                if len(token) == len(merged[start:i+1]):
                    possible_choices.append(token)
        if not available_tokens or i==len(merged)-1:
            possible_choices.sort(reverse=True, key= lambda x:len(x))
            if possible_choices:
                splited_tokens.append(possible_choices[0])
                start = start + len(possible_choices[0])
            available_tokens = tokens
            possible_choices.clear()
    if sum(list(map(lambda x:len(x),splited_tokens))) != len(merged):
        return 0,splited_tokens
    else:
        return 1,splited_tokens
```

```python
eng_tokens = open('en.tokens.en','r')
eng_tokens = eng_tokens.readlines()
eng_tokens = list(map(lambda x : x[:-1], eng_tokens))
eng_merged_words = open('mergedTokens.en')
eng_merged_words = eng_merged_words.readlines()
eng_merged_words = list(map (lambda x:x[0:-1], eng_merged_words))
```

```python
fa_tokens = open('fa.words.txt','r')
fa_tokens = fa_tokens.readlines()
fa_tokens = list(map(lambda x : x[:-1], fa_tokens))
fa_merged_words = open('mergedTokens.fa')
fa_merged_words = fa_merged_words.readlines()
fa_merged_words = list(map(lambda x : x[0:-1], fa_merged_words))
fa_tokens = list(map(lambda x : x.split("\t")[0],fa_tokens))
fa_tokens[0]='و'
```

```python
fa_responses = []
en_responses = []
```

```python
def divider(arr, size):
    batches = []
    print(len(arr))
    n = len(arr)
    i=0
    while(n>=size):
        batches.append(arr[i*size:(i+1)*size+1])
        i=i+1
        n-=size
    if(n>0):
        batches.append(arr[i*n:])
    return batches

def fa_tokenizer(arr):
    for word in arr:
        res,tokens = max_length_tokenizer(word,fa_tokens)
        if res == 0 :
            response = word +" Error" + str(tokens)[1:-1]
        else: 
            response = word +" Success" + str(tokens)[1:-1]
        fa_responses.append(response)
       
def en_tokenizer(arr):
    for word in arr:
        res,tokens = max_length_tokenizer(word,eng_tokens)
        if res == 0 :
            response = word +" Error" + str(tokens)[1:-1]
        else: 
            response = word +" Success" + str(tokens)[1:-1]
        en_responses.append(response)


        
def parallelizer(merged_arr, lang, threads_number):
    size = int(len(merged_arr)/threads_number)
    batches = divider(merged_arr,size)
    print(f'{threads_number} Threads will be using for processing {len(merged_arr)} which has divided into {len(merged_arr)/size} arrays')
    threads = []
    if lang == 'fa':
        fa_responses.clear()
        target = fa_tokenizer
    else:
        en_responses.clear()
        target = en_tokenizer
    
    for i in range(len(batches)):
        thread = threading.Thread(target = target,args = [batches[i]])
        threads.append(thread)
        print(f'worker {i} is starting')
        thread.start()
    for t in threads:
        t.join()
        print(f'worker has finished its work')
        
    
```
